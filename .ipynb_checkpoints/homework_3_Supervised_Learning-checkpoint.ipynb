* Импортируйте библиотеки pandas и numpy.


* Загрузите "Boston House Prices dataset" из встроенных наборов данных библиотеки sklearn. Создайте датафреймы X и y из этих данных.


* Разбейте эти датафреймы на тренировочные (X_train, y_train) и тестовые (X_test, y_test) с помощью функции train_test_split так, чтобы размер тестовой выборки составлял 30% от всех данных, при этом аргумент random_state должен быть равен 42.


* Создайте модель линейной регрессии под названием lr с помощью класса LinearRegression из модуля sklearn.linear_model.


* Обучите модель на тренировочных данных (используйте все признаки) и сделайте предсказание на тестовых.

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
from sklearn.datasets import load_boston # из библеотеки склерн импортируем дадасет с данными по нед. Бостона
```

```python
boston = load_boston() # загружаем словарь данных в переменную boston

boston.keys() # Смотрим какие есть ключи у словаря
```

```python
data = boston["data"] # Данные о недвижимости хранятся в массиве по ключу "data". Посмотрим поближе на эти данные.

data.shape
```

```python
feature_names = boston["feature_names"] # столбцы - характеристики. Названия этих признаков хранятся в массиве по ключу "feature_names"

feature_names 
```

```python
print(boston["DESCR"]) # описание признаков и информация об авторах 
```

```python
from sklearn.model_selection import train_test_split # импотрируем функцию создания сплита
```

```python
target = boston["target"] # массив с целевыми значениями (в нашем случае - ценами на недвижимость)

target[:10] # первые 10 целевых значений
```

```python
# В таблице X будут храниться признаки. В качестве названий для столбцов возьмём массив feature_names
X = pd.DataFrame(data, columns = feature_names) 

X.head() # первые 5 строк
```

```python
X.info() # Выведем информацию об этой таблице. В каждом столбце по 506 строк из 506. Пропусков нет. 
```

```python
y = pd.DataFrame(target, columns=["price"]) # Теперь создадим таблицу y, в которую запишем целевые значения

y.info()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 42) 
# Разбиение выборки на тренировочную и тестовую test_size=0.3 - 30% выборки 
# X_train & y_train - тренировочные переменные 
# X_test & y_test - переменные для проверки модели
```

```python
from sklearn.linear_model import LinearRegression # Загружаем модель линейной регрессии 
```

```python
lr = LinearRegression() # погружаем ее в переменную
```

```python
lr.fit(X_train, y_train) # метод fit тренирует модель на данных которые мы отправили (1 признаки, 2 цель)
```

```python
y_pred = lr.predict(X_test) # Метод predict теперь уже на тестовых данных предсказывает целевую переменную y_pred

y_pred.shape # как мы видим, это - двумерный массив
```

```python
# y_test реальные цены 
# y_pred предсказанные цены 
check_test = pd.DataFrame({
    "y_test": y_test["price"],
    "y_pred": y_pred.flatten(),
})

check_test.head(10)
```

```python
# смотрим величину ошибки 
check_test["error"] = check_test["y_pred"] - check_test["y_test"]

check_test.head()
```

```python
from sklearn.metrics import r2_score
```

```python
# оцениваем модель с помощью квадрата ошибки 0 < оштбки < 1 
r2_score(y_test, y_pred) 
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv('creditcard.csv')
df
```

```python
df["Class"].value_counts(normalize = True)
```

```python
df.info()
```

```python
pd.options.display.max_columns = 100
```

```python
df.head(10)
```

```python
target = 'Class'
```

```python
y = df[target]
y
```

```python
X = df.drop(target, axis = 1)
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 100, stratify = y) 
```

```python
print(f"Размер тренировочной матрицы: {X_train.shape}")
print(f"Размер целевой переменной: {y_train.shape}")
print(f"Размер тестовой матрицы: {X_test.shape}")
print(f"Размер тестовой переменной: {y_test.shape}")
```

```python
parameters = [{'n_estimators': [10, 15],
            'max_features': np.arange(3, 5),
            'max_depth': np.arange(4, 7)}]
```

```python
clf = GridSearchCV(
    estimator = RandomForestClassifier(random_state = 100),
    param_grid = parameters,
    scoring = 'roc_auc',
    cv = 3
)
```

```python
%%time
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
y_pred = clf.predict_proba(X_test)
y_pred
```

```python
y_pred_proba = y_pred[:, 1]
y_pred_proba
```

```python
# Почему мы всегда так иделально угадываем клас 0? 

plt.hist(y_pred_proba[y_test == 1], bins=4, density=True, label='class 1', alpha=0.5)
plt.hist(y_pred_proba[y_test == 0], bins=4, density=True, label='class 0', alpha=0.5)

plt.xlabel("Probability")
plt.ylabel("Density")

plt.grid(linestyle = '--', alpha=0.5)

plt.legend()
```

```python
from sklearn.metrics import roc_curve
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba, pos_label = 1)
```

```python
plt.rcParams['figure.figsize'] = 5, 5

plt.plot(fpr, tpr)
plt.plot([0, 1], [0, 1], color = 'grey', linestyle = 'dashed')

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC curve')

plt.grid(linestyle = '--', alpha = 0.5)

plt.show()
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(y_test, y_pred_proba) 
```
